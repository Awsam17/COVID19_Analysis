* Import needed modules :

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('covid19_Confirmed_dataset.csv')
```

* Check the dataset format :

```python
dataset.head()
```

```python
dataset.shape
```

* Delete useless columns :

```python
dataset.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
dataset_aggregated.head(20)
```

```python
dataset_aggregated.drop(["Province/State"],axis=1,inplace=True)
```

```python
dataset_aggregated.head()
```

```python
dataset_aggregated.shape
```

* Visualize data for certain country :

```python
dataset_aggregated.loc["China"].plot()
dataset_aggregated.loc["Albania"].plot()
plt.legend()
```

```python
dataset_aggregated.loc["China"][:3].plot()
```

* Measure the spread of the virus :

```python
dataset_aggregated.loc["China"].diff().plot()
```

* Maximum infection rate :

```python
dataset_aggregated.loc["China"].diff().max()
```

```python
dataset_aggregated.loc["Italy"].diff().max()
```

```python
dataset_aggregated.loc["Spain"].diff().max()
```

* Max infection rate of all countries

```python
countries = list(dataset_aggregated.index)
```

```python
max_infection_rate = []
for c in countries :
  max_infection_rate.append(dataset_aggregated.loc[c].diff().max())
```

```python
max_infection_rate
```

```python
dataset_aggregated["max_infection_rate"] = max_infection_rate
```

```python
dataset_aggregated.head()
```

```python
corona_dataset = pd.DataFrame(dataset_aggregated["max_infection_rate"])
corona_dataset.head()
```

* Analyze another related dataset :

```python
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
```

* Check the dataset :

```python
happiness_report.head()
```

```python
happiness_report.shape
```

* Drop useless columns :

```python
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.head()
```

```python
happiness_report_agg = happiness_report.groupby("Country or region").sum()
```

```python
dataset_aggregated.shape
```

```python
final_dataset = corona_dataset.join(happiness_report_agg,how="inner")
final_dataset.head()
```

```python
final_dataset.shape
```

* Correlation matrix :

```python
final_dataset.corr()
```

* GDP vs Max infection rate :

```python
x = final_dataset["GDP per capita"]
y = final_dataset["max_infection_rate"]
sns.scatterplot(x=x,y=np.log(y))
# to solve high scaling difference between x and y use log to the highest scaling axis
```

```python
sns.regplot(x=x,y=np.log(y))
```
